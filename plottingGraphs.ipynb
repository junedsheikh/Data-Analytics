```python
## Few concepts here

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from string import digits

amazon = pd.read_csv("amazon_jobs_dataset.csv")

df = amazon.copy()

date = df["Posting_date"]
date = date.str.split(' ', expand = True)  ## This split() with expand = True expands the string into 
                                           ## seprate columns.

month = date[0]
freq = month.value_counts()
months = freq.index
jobs = freq.values

plt.bar(months, jobs)
plt.xticks(rotation = 40)
plt.show()

for i in range(len(months)):
    print(months[i], jobs[i])
```

```python
import numpy as np
import csv
import matplotlib.pyplot as plt
file=pd.read_csv("amazon_jobs_dataset.csv")
file_data=csv.DictReader(file,skipinitialspace=True)
data={}
array1=[]
array2=[]
for ele in file_data:
    city=ele["location"].str.split('',expand=True)
    if len(city)>1 and city[0]=="IN,":
        
        if city[2] in data:
            data[city[2]]+=1
        else:
            data[city[2]]=1
                
for ele in data:
    array1.append(ele)
    array2.append(data[ele])
        
        # text=str(ele)+","+str(data[ele])
        # plt.text(ele+0.1,data[ele]+0.1,text)
plt.pie(array2,labels=array1,autopct="%.2f")
plt.show()
final=np.array(array2)
for ele in data:
    percentage=data[ele]*100/final.sum()
    print(ele,"%.2f"%percentage,sep=" ")
   
```

```python
#Project : Case Study Investment
#Number of Fundings

import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv("startup_funding.csv")
dates = data['Date'].str[-4:]

d = {}
for i in dates:
    d[i] = d.get(i,0) + 1
        
values = []
quantity = []
for i in sorted(d.keys()):
    values.append(i)
    quantity.append(d[i])
    
plt.plot(values,quantity,color = "darkblue", marker = "s")
plt.grid()
plt.show()

for i in range(len(values)):
    print(values[i],quantity[i])
```

```python
import matplotlib.pyplot as plt
import pandas as pd

df = pd.read_csv('startup_funding.csv')
cities = df['CityLocation']
df['CityLocation'].dropna(inplace = True)
df['CityLocation'].replace('Delhi','New Delhi', inplace = True)
df['CityLocation'].replace('bangalore','Bangalore', inplace = True)

def valid_city(city):
    city = str(city)
    if city != '':
        return city.split('/')[0].strip()
    
cities = cities.apply(valid_city)
startup = cities.value_counts()

x = startup.index[:10]
y = startup[:10]

plt.pie(y, labels = x)
plt.title("PIE-CHART", color = 'navy', fontsize = 18)
plt.show()

for i in range(len(x)):
    print(x[i],y[i])
```

```python
#Funding amount

import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

df = pd.read_csv("startup_funding.csv")
#city = df['CityLocation']

df['CityLocation'].fillna("Unknown",inplace = True)
df['CityLocation'].replace("Delhi","New Delhi",inplace = True)
df['CityLocation'].replace("bangalore","Bangalore",inplace = True)

def seperatecity(city):
        return city.split("/")[0].strip()
    
df['CityLocation'] = df['CityLocation'].apply(lambda x: seperatecity(x))
df['AmountInUSD'] = df['AmountInUSD'].apply(lambda x: float(str(x).replace(",","")))
df['AmountInUSD'] = pd.to_numeric(df['AmountInUSD'])
investment_amt = df[df['CityLocation']!="Unknown"].groupby('CityLocation')['AmountInUSD'].sum().sort_values(ascending=False)
cities = investment_amt.index[:10]
investment_val = investment_amt.values[:10]
per = np.true_divide(investment_val,investment_val.sum())*100
y=investment_val[:10]
for i in range(len(cities)):
    print(cities[i],format(per[i],'.2f'))
```

```python
#Investment Type

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

df_start=pd.read_csv('startup_funding.csv',encoding='utf8')
df_start['InvestmentType'].dropna(inplace=True)

def separateCity(city):
    return city.split('/')[0].strip()

df_start['InvestmentType']=df_start['InvestmentType'].apply(lambda x:separateCity)
df_start['InvestmentType'].replace("PrivateEquity","Private Equity",inplace=True)
df_start['InvestmentType'].replace("SeedFunding","Seed Funding",inplace=True)
df_start['InvestmentType'].replace("Crowd funding","Crowd Funding",inplace=True)

## Converting "AmountInUSD" into numeric format
df_start["AmountInUSD"] = df_start["AmountInUSD"].apply(lambda x: float(str(x).replace(",","")))
df_start["AmountInUSD"] = pd.to_numeric(df_start["AmountInUSD"])
city_amount=df_start.groupby('InvestmentType')['AmountInUSD'].sum().sort_values(ascending=False)[0:10]
city=city_amount.index
amountCity=city_amount.values
perAmount=np.true_divide(amountCity, amountCity.sum())*100
plt.pie(perAmount,labels = city , autopct = "%.2f%%")
plt.axis("equal")
plt.show()
for i in range(len(city)):
    print(city[i],format(perAmount[i],'.2f'))
```
